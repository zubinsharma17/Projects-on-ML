```python
import pandas as pd
import matplotlib.pyplot as mt
import numpy as np

data=pd.read_csv("quicker.csv")
data
```

```python

data.head()
```

```python
data['Price'].isnull()
data['name']
```

```python
data["Price"].nunique
data1=data.copy()

data1["Price"]=data1["Price"].str.replace(',','')
data1["Price"]

data1=data1[data1['Price'].str.isnumeric()]
data1['Price']=data1['Price'].astype(int)
data1.info()
```

```python
data1=data1[data1['year'].str.isnumeric()]
data1['year']=data1['year'].astype(int)
data1.info()
data1['year']
data1['name'].str.split(' ').str.slice(0,3).str.join(" ")
data1['kms_driven']=data1['kms_driven'].str.split(" ").str.get(0).str.replace(',','')
data1=data1[data1['kms_driven'].str.isnumeric()]
data1['kms_driven'].unique()
mt.scatter(data1['Price'], data1['year']);
mt.show()
```

```python
data1['kms_driven']=data1['kms_driven'].astype(int)
data1.info()
data1.to_csv('cleaned_car.csv')
x = data1.drop(columns='Price')
y=data1['Price']
```

```python
x
```

```python
y
```

```python

from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score

```

```python
ohe=OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
ohe.categories_

```

```python
ct=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

```python
lr=LinearRegression()

```

```python
pipe=make_pipeline(ct,lr)
pipe.fit(x_train,y_train)
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(ct,lr)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
np.argmax(scores)
scores[np.argmax(scores)]
pipe.predict(pd.DataFrame(columns=x_test.columns,data=np.array(['Maruti Suzuki Swift','Maruti',2019,100,'Petrol']).reshape(1,5)))

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(ct,lr)
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
r2_score(y_test,y_pred)
```
